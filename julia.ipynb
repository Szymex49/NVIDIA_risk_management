```julia
using Plots
using StatsBase
using Distributions
using StatsFuns
using KernelDensity
using Statistics
using StatsPlots
using CSV
using DataFrames
using LaTeXStrings
using LsqFit
using DelimitedFiles
```

```julia
function regression(X, Y, display_plot=true)
    r = cor(X, Y)
    Sx = std(X)
    Sy = std(Y)
    a = r * Sy / Sx
    b = mean(Y) - a * mean(X)

    if display_plot
        scatter(X, Y, ms=2, markerstrokewidth=0.5, alpha=0.5)
        xs = minimum(X):0.01:maximum(X)
        plot!(xs, a .* xs .+ b, c=:red, lw=2) |> display
    end

    return [a, b]
end
```

```julia
X = rand(Normal(0, 1), 1000)
Y = X .+ rand(Normal(0, 1), 1000)
regression(X, Y)
```
