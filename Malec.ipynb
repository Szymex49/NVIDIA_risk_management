```r
library(dplyr)
library(ggplot2)
library(zeallot)
library(tidyr)
library(reshape2)
library(cowplot)
library(latex2exp)
library(stats)
library(aTSA)
options(repr.plot.width = 12, repr.plot.height = 8)
```

```r
copper <- read.csv('data/copper.csv', header=TRUE)
copper %>% head
```

```r
X <- copper$Open
```

```r
ggplot() + geom_line(aes(1:length(X), X))
```

```r
X_diff <- diff(X)
ggplot() + geom_line(aes(1:length(X_diff), X_diff))
```
