```r
library(dplyr)
library(ggplot2)
library(zeallot)
library(tidyr)
library(reshape2)
library(cowplot)
library(latex2exp)
library(stats)
library(aTSA)
options(repr.plot.width = 12, repr.plot.height = 8)
```

```r
copper <- read.csv('data/copper.csv', header=TRUE)
copper %>% head
```

```r
X <- copper$Open
```

```r
ggplot() + geom_line(aes(1:length(X), X))
```

```r
X_diff <- diff(X)
ggplot() + geom_line(aes(1:length(X_diff), X_diff))
```

```r
Y <- log(X[2:length(X)] / X[1:length(X)-1])
ggplot() + geom_line(aes(1:length(Y), Y))
```

```r
Y <- log(X[2:length(X)] / X[1:length(X)-1])
Y <- Y / sd(Y)
xs <- seq(-5, 5, 0.01)
ggplot() + geom_histogram(aes(x=Y, y=..density..), bins=50) + geom_line(aes(xs, dnorm(xs)))
```
